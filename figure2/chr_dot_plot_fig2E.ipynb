### Description:
This script parses a reference genome in FASTA format and loads it into a dictionary. It opens the reference genome file, reads each sequence using `SeqIO.parse`, and stores the sequences in the `assembly` dictionary with chromosome names as keys. It also creates a dictionary `assembly_sequence_length` to store the length of each chromosome sequence. Finally, it closes the reference genome file. The resulting dictionaries are prepared for further analysis, with `assembly` containing the full sequences and `assembly_sequence_length` containing the corresponding lengths of each chromosome.


````python
from Bio import SeqIO
#parse CDR based on A count 


assembly_ = open("/private/groups/migalab/dan/reference/hg002v1.0.1.fasta", "r")

#Load the reference genome and make it into a dictionary 
fasta_sequences = SeqIO.parse(assembly_, "fasta")
assembly={}
for fasta in fasta_sequences:
    name, sequence = fasta.id, str(fasta.seq)
    assembly[name] = sequence

#Make a dictionary for all the chromosomes and their corresponding sequence length 
assembly_sequence_length = {}    
for chromosome in assembly:
    assembly_sequence_length[chromosome] = len(assembly[chromosome])
    
assembly_.close()

````

### Description:
The `comprehensive_mA_density_calculation` function calculates the comprehensive mA density for given regions in paternal and maternal chromosomes. It takes a DataFrame containing region data and a dictionary with chromosome sequences. For each chromosome (from chr1 to chr22), it calculates the total mA density by counting the number of 'A' bases in the specified regions and multiplying it by the density values from the DataFrame. The function computes the comprehensive density for both paternal and maternal chromosomes and averages them. It also calculates the percentage difference in density between the two and stores the results in two DataFrames: `region_weighted_df` (containing the differences and averages) and `region_combined_df` (containing the comprehensive densities). The function returns these two DataFrames.

The function is applied to three datasets: `CENPA_CDR_df`, `CENPA_chrom_arm_df`, and `CENPA_non_active_df`, producing corresponding weighted and combined DataFrames for each dataset.


````python
import ast
import pandas as pd

def comprehensive_mA_density_calculation(region_df, assembly, bed_file_path):   
    # Load the bed file to get desired read lengths
    bed_df = pd.read_csv(bed_file_path, sep='\t', header=None)
    
    # Determine number of columns and assign names
    num_cols = len(bed_df.columns)
    
    if num_cols == 3:
        bed_df.columns = ['Chromosome', 'start', 'end']
    elif num_cols == 9:
        bed_df.columns = ['Chromosome', 'start', 'end', 'name', 'score', 'strand', 'thickStart', 'thickEnd', 'itemRgb']
    else:
        bed_df.columns = [f'col{i}' for i in range(num_cols)]
        bed_df.rename(columns={'col0': 'Chromosome', 'col1': 'start', 'col2': 'end'}, inplace=True)
    
    # Calculate total length per chromosome
    bed_df['region_length'] = bed_df['end'] - bed_df['start']
    total_length_per_chr = bed_df.groupby('Chromosome')['region_length'].sum().to_dict()
    
    region_weighted_df = pd.DataFrame(columns=['Chromosome', 'Difference', 'Average'])
    region_combined_df = pd.DataFrame(columns=['Chromosome', 'Start/End', 'Density', 'Desired_Read_Length'])

    # Filter rows for chromosome 1 to 22
    chromosomes = [f'chr{i}' for i in range(1, 23)]

    for chrom in chromosomes:

        maternal_col = f'{chrom}_MATERNAL'
        paternal_col = f'{chrom}_PATERNAL'

        #grab chromosome hap coordinates and put them in seperate lists 
        MAT_indices = region_df.index[region_df['Chromosome'] == maternal_col].tolist()
        PAT_indices = region_df.index[region_df['Chromosome'] == paternal_col].tolist()

        #make the start and end positions of each and every chromosome into and individual list 
        PAT_CDR_coordinates_from_density = region_df.loc[PAT_indices]['Start/End'].tolist()
        MAT_CDR_coordinates_from_density = region_df.loc[MAT_indices]['Start/End'].tolist()
        
        PAT_CDR_coordinates_total_A = 0 
        PAT_CDR_coordinates_total_mA = 0 

        # calculate the total amount of mA from the total amount of As in the reference genome and density 
        # calculated in the previous script in both Paternal and Maternal chromosomes 
        for coordinate in range (len(PAT_CDR_coordinates_from_density)): 
            
            PAT_CDR_coordinates_start_end = ast.literal_eval(PAT_CDR_coordinates_from_density[coordinate])
            PAT_CDR_coordinates_start = int (PAT_CDR_coordinates_start_end[0])
            PAT_CDR_coordinates_end = int (PAT_CDR_coordinates_start_end[1])
            PAT_CDR_A_count = assembly[paternal_col][PAT_CDR_coordinates_start:PAT_CDR_coordinates_end].count('A')
            PAT_CDR_coordinates_total_A += PAT_CDR_A_count
            PAT_CDR_coordinates_total_mA += PAT_CDR_A_count * region_df['Density'][PAT_indices[coordinate]]

        

        try:
            PAT_CDR_coordinates_comprehensive_density = PAT_CDR_coordinates_total_mA / PAT_CDR_coordinates_total_A
        except ZeroDivisionError:
            pass 
        new_row = pd.DataFrame({
                                    'Chromosome': [paternal_col],
                                    'Start/End': [[PAT_CDR_coordinates_start, PAT_CDR_coordinates_end]],
                                    'Density': [PAT_CDR_coordinates_comprehensive_density],
                                    'Desired_Read_Length': [total_length_per_chr.get(paternal_col, 0)]
                                })
        

        region_combined_df = pd.concat([region_combined_df, new_row], ignore_index=True)


        MAT_CDR_coordinates_total_A = 0 
        MAT_CDR_coordinates_total_mA = 0 

        for coordinate in range (len(MAT_CDR_coordinates_from_density)): 
            MAT_CDR_coordinates_start_end = ast.literal_eval(MAT_CDR_coordinates_from_density[coordinate])
            MAT_CDR_coordinates_start = int (MAT_CDR_coordinates_start_end[0])
            MAT_CDR_coordinates_end = int (MAT_CDR_coordinates_start_end[1])
            MAT_CDR_A_count = assembly[maternal_col][MAT_CDR_coordinates_start:MAT_CDR_coordinates_end].count('A')
            MAT_CDR_coordinates_total_A += MAT_CDR_A_count
            MAT_CDR_coordinates_total_mA += MAT_CDR_A_count * region_df['Density'][MAT_indices[coordinate]]

        try:
            MAT_CDR_coordinates_comprehensive_density = MAT_CDR_coordinates_total_mA / MAT_CDR_coordinates_total_A
        except ZeroDivisionError:
            pass 
        new_row = pd.DataFrame({
                                    'Chromosome': [maternal_col],
                                    'Start/End': [[MAT_CDR_coordinates_start, MAT_CDR_coordinates_end]],
                                    'Density': [MAT_CDR_coordinates_comprehensive_density],
                                    'Desired_Read_Length': [total_length_per_chr.get(maternal_col, 0)]
                                })
        region_combined_df = pd.concat([region_combined_df, new_row], ignore_index=True)


        chromosome_average = (PAT_CDR_coordinates_comprehensive_density + MAT_CDR_coordinates_comprehensive_density) / 2


        diff = abs (((PAT_CDR_coordinates_comprehensive_density - MAT_CDR_coordinates_comprehensive_density)/ MAT_CDR_coordinates_comprehensive_density)*100)
        regional_new_row = pd.DataFrame({
                                    'Chromosome': [chrom],
                                    'Difference': [diff],
                                    'Average': [chromosome_average]
                                })
        
        region_weighted_df = pd.concat([region_weighted_df, regional_new_row], ignore_index=True)
    return region_weighted_df, region_combined_df

CENPA_CDR_A_weighted_young_df, CENPA_CDR_A_combined_young_df = comprehensive_mA_density_calculation(CENPA_CDR_young_df, assembly, '/private/groups/migalab/dan/data_analysis/young_old_analysis/HG002_DiMeLo_CENPA_youngpassage.hmmCDR_only_CDR_Dan_certified.bed')
CENPA_CDR_A_weighted_old_df, CENPA_CDR_A_combined_old_df = comprehensive_mA_density_calculation(CENPA_CDR_old_df, assembly, '/private/groups/migalab/dan/data_analysis/young_old_analysis/HG002_DiMeLo_CENPA_oldpassage.hmmCDR_only_CDR_Dan_certified.bed')
CENPC_CDR_A_weighted_young_df, CENPC_CDR_A_combined_young_df = comprehensive_mA_density_calculation(CENPC_CDR_young_df, assembly, '/private/groups/migalab/dan/data_analysis/young_old_analysis/HG002_DiMeLo_CENPA_youngpassage.hmmCDR_only_CDR_Dan_certified.bed')
CENPA_CDR_A_weighted_iPSC_df, CENPA_CDR_A_combined_iPSC_df = comprehensive_mA_density_calculation(CENPA_CDR_iPSC_df, assembly, '/private/groups/migalab/dan/data_analysis/hiPSC_CDR_included_domains.bed')

#CENPA_chrom_arm_weighted_df, CENPA_chrom_arm_A_combined_df = comprehensive_mA_density_calculation(CENPA_chrom_arm_df, assembly, 'path/to/CENPA_chrom_arm.bed')
#CENPA_non_active_weighted_df, CENPA_non_active_A_combined_df = comprehensive_mA_density_calculation(CENPA_non_active_df, assembly, 'path/to/CENPA_non_active.bed')
````

````python
import ast
import numpy as np
import re

def _parse_start_end(val):
    """Return (start, end) as ints from a list/tuple or a string like '[123, 456]'."""
    if isinstance(val, (list, tuple)) and len(val) >= 2:
        return int(val[0]), int(val[1])
    if isinstance(val, str):
        try:
            parsed = ast.literal_eval(val)
            if isinstance(parsed, (list, tuple)) and len(parsed) >= 2:
                return int(parsed[0]), int(parsed[1])
        except Exception:
            pass
        parts = re.findall(r"-?\d+", val)
        if len(parts) >= 2:
            return int(parts[0]), int(parts[1])
    raise ValueError(f"Unrecognized Start/End format: {val!r}")

def _count_As_in_region(row, asm_dict):
    chrom = row['Chromosome']
    start, end = _parse_start_end(row['Start/End'])
    seq = asm_dict.get(chrom)
    if seq is None:
        return np.nan
    s, e = sorted((max(0, start), min(len(seq), end)))
    if e <= s:
        return 0
    return seq[s:e].count('A')

def add_A_count_and_mA_estimate(combined_df, asm_dict):
    """
    Adds 'A_count' and 'mA_estimate' columns to a combined density DataFrame.
    Desired_Read_Length is already correctly summed from the BED file
    in comprehensive_mA_density_calculation and must NOT be overwritten here.

    Parameters:
        combined_df : pd.DataFrame with columns ['Chromosome', 'Start/End', 'Density', 'Desired_Read_Length']
        asm_dict    : dict mapping chromosome names to sequence strings (i.e. assembly)

    Returns:
        combined_df with new columns added (also returns for convenience).
    """
    combined_df['A_count'] = combined_df.apply(
        _count_As_in_region, axis=1, asm_dict=asm_dict
    )
    combined_df['mA_estimate'] = (
        combined_df['Density'].astype(float).fillna(0.0)
        * combined_df['A_count'].astype(float).fillna(0.0)
    )
    return combined_df


CENPA_CDR_A_combined_young_df = add_A_count_and_mA_estimate(CENPA_CDR_A_combined_young_df, assembly)
CENPA_CDR_A_combined_old_df   = add_A_count_and_mA_estimate(CENPA_CDR_A_combined_old_df,   assembly)
CENPC_CDR_A_combined_young_df = add_A_count_and_mA_estimate(CENPC_CDR_A_combined_young_df, assembly)
CENPA_CDR_A_combined_iPSC_df  = add_A_count_and_mA_estimate(CENPA_CDR_A_combined_iPSC_df,  assembly)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import ast

def plot_cenpa_dosage(df_combined, df_cdr,
                     xmin_density=0, xmax_density=0.01,
                     ymin_density=0, ymax_density=0.01,
                     xmin_len=0,     xmax_len=None,
                     ymin_len=0,     ymax_len=None,
                     xmin_product=0, xmax_product=None,
                     ymin_product=0, ymax_product=None,
                     z_thresh=3.0,   save_path=None):

    # ====== Step 1: Get Desired_Read_Length directly from df_combined ======
    pat_len = df_combined[df_combined['Chromosome'].str.contains('PATERNAL')][['Chromosome', 'Desired_Read_Length']].copy()
    mat_len = df_combined[df_combined['Chromosome'].str.contains('MATERNAL')][['Chromosome', 'Desired_Read_Length']].copy()
    pat_len['Chromosome_Label'] = pat_len['Chromosome'].str.replace('_PATERNAL', '', regex=True)
    mat_len['Chromosome_Label'] = mat_len['Chromosome'].str.replace('_MATERNAL', '', regex=True)

    # ====== Step 2: Get density from df_combined ======
    paternal_df = df_combined[df_combined['Chromosome'].str.contains('PATERNAL')].copy()
    maternal_df = df_combined[df_combined['Chromosome'].str.contains('MATERNAL')].copy()
    paternal_df['Chromosome_Label'] = paternal_df['Chromosome'].str.replace('_PATERNAL', '', regex=True)
    maternal_df['Chromosome_Label'] = maternal_df['Chromosome'].str.replace('_MATERNAL', '', regex=True)

    # ====== Step 3: Merge density + length ======
    merged_df = pd.merge(
        paternal_df[['Chromosome_Label', 'Density']],
        maternal_df[['Chromosome_Label', 'Density']],
        on='Chromosome_Label', suffixes=('_PATERNAL', '_MATERNAL')
    )
    merged_df = merged_df.merge(
        pat_len[['Chromosome_Label', 'Desired_Read_Length']].rename(columns={'Desired_Read_Length': 'Total_Length_PATERNAL'}),
        on='Chromosome_Label'
    )
    merged_df = merged_df.merge(
        mat_len[['Chromosome_Label', 'Desired_Read_Length']].rename(columns={'Desired_Read_Length': 'Total_Length_MATERNAL'}),
        on='Chromosome_Label'
    )
    merged_df['Short_Label'] = merged_df['Chromosome_Label'].str.replace(r'^chr', '', regex=True)

    # Force numeric
    merged_df['Total_Length_PATERNAL'] = pd.to_numeric(merged_df['Total_Length_PATERNAL'], errors='coerce').astype(float)
    merged_df['Total_Length_MATERNAL'] = pd.to_numeric(merged_df['Total_Length_MATERNAL'], errors='coerce').astype(float)
    merged_df['Density_PATERNAL'] = pd.to_numeric(merged_df['Density_PATERNAL'], errors='coerce').astype(float)
    merged_df['Density_MATERNAL'] = pd.to_numeric(merged_df['Density_MATERNAL'], errors='coerce').astype(float)

    # ====== Step 4: Compute product columns ======
    merged_df['Product_PATERNAL'] = merged_df['Density_PATERNAL'] * merged_df['Total_Length_PATERNAL']
    merged_df['Product_MATERNAL'] = merged_df['Density_MATERNAL'] * merged_df['Total_Length_MATERNAL']

    # ====== Print data ranges ======
    print("===== Data Ranges =====")
    print(f"Density PATERNAL: {merged_df['Density_PATERNAL'].min():.6f} to {merged_df['Density_PATERNAL'].max():.6f}")
    print(f"Density MATERNAL: {merged_df['Density_MATERNAL'].min():.6f} to {merged_df['Density_MATERNAL'].max():.6f}")
    print(f"Length  PATERNAL: {merged_df['Total_Length_PATERNAL'].min()} to {merged_df['Total_Length_PATERNAL'].max()}")
    print(f"Length  MATERNAL: {merged_df['Total_Length_MATERNAL'].min()} to {merged_df['Total_Length_MATERNAL'].max()}")
    print(f"Product PATERNAL: {merged_df['Product_PATERNAL'].min():.2f} to {merged_df['Product_PATERNAL'].max():.2f}")
    print(f"Product MATERNAL: {merged_df['Product_MATERNAL'].min():.2f} to {merged_df['Product_MATERNAL'].max():.2f}")
    print("\nPer-chromosome lengths:")
    print(merged_df[['Chromosome_Label', 'Total_Length_PATERNAL', 'Total_Length_MATERNAL']].to_string())

    # ====== Auto limits ======
    if xmax_len is None:
        xmax_len = merged_df['Total_Length_PATERNAL'].max() * 1.05
    if ymax_len is None:
        ymax_len = merged_df['Total_Length_MATERNAL'].max() * 1.05
    if xmax_product is None:
        xmax_product = merged_df['Product_PATERNAL'].max() * 1.05
    if ymax_product is None:
        ymax_product = merged_df['Product_MATERNAL'].max() * 1.05

    # ====== Colors ======
    unique_chrs = merged_df['Chromosome_Label'].unique()
    palette = sns.color_palette("tab20", len(unique_chrs))
    color_map = dict(zip(unique_chrs, palette))

    # ====== Helper functions ======
    def compute_stats(y, x):
        y = pd.to_numeric(y, errors='coerce').astype(float).values
        x = pd.to_numeric(x, errors='coerce').astype(float).values
        r = np.corrcoef(y, x)[0, 1]
        mse = np.mean((y - x) ** 2)
        return r, mse

    def get_outliers(y, x):
        y = pd.to_numeric(y, errors='coerce').astype(float).values
        x = pd.to_numeric(x, errors='coerce').astype(float).values
        resid = y - x
        mu = float(np.mean(resid))
        sd = float(np.std(resid, ddof=1))
        if sd > 0:
            z = np.abs((resid - mu) / sd)
            return z > z_thresh, mu, sd
        else:
            return np.array([False] * len(y)), mu, sd

    # ====== Stats & outliers ======
    r_dens, mse_dens = compute_stats(merged_df['Density_MATERNAL'], merged_df['Density_PATERNAL'])
    outlier_dens, mu_dens, sd_dens = get_outliers(merged_df['Density_MATERNAL'], merged_df['Density_PATERNAL'])
    merged_df['is_outlier_density'] = outlier_dens

    r_len, mse_len = compute_stats(merged_df['Total_Length_MATERNAL'], merged_df['Total_Length_PATERNAL'])
    outlier_len, mu_len, sd_len = get_outliers(merged_df['Total_Length_MATERNAL'], merged_df['Total_Length_PATERNAL'])
    merged_df['is_outlier_length'] = outlier_len

    r_prod, mse_prod = compute_stats(merged_df['Product_MATERNAL'], merged_df['Product_PATERNAL'])
    outlier_prod, mu_prod, sd_prod = get_outliers(merged_df['Product_MATERNAL'], merged_df['Product_PATERNAL'])
    merged_df['is_outlier_product'] = outlier_prod

    print("\n===== Density =====")
    print(f"r = {r_dens:.4f}, MSE = {mse_dens:.6f}")
    print(f"mu = {mu_dens:.6f}, sd = {sd_dens:.6f}")
    print("Outliers:", merged_df.loc[outlier_dens, 'Chromosome_Label'].tolist() if outlier_dens.sum() > 0 else "None")

    print("\n===== Length =====")
    print(f"r = {r_len:.4f}, MSE = {mse_len:.6f}")
    print(f"mu = {mu_len:.2f}, sd = {sd_len:.2f}")
    print("Outliers:", merged_df.loc[outlier_len, 'Chromosome_Label'].tolist() if outlier_len.sum() > 0 else "None")

    print("\n===== Product (Density x Length) =====")
    print(f"r = {r_prod:.4f}, MSE = {mse_prod:.6f}")
    print(f"mu = {mu_prod:.4f}, sd = {sd_prod:.4f}")
    print("Outliers:", merged_df.loc[outlier_prod, 'Chromosome_Label'].tolist() if outlier_prod.sum() > 0 else "None")

    # ====== Inner plotting function ======
    def make_plot(ax, x_col, y_col, outlier_col, sd, xlabel, ylabel,
                  xmin, xmax, ymin, ymax):
        for _, row in merged_df.iterrows():
            fontsize = 14 if row[outlier_col] else 12
            ax.text(
                row[x_col],
                row[y_col],
                row['Short_Label'],
                fontsize=fontsize,
                ha='center',
                va='center',
                color=color_map[row['Chromosome_Label']],
                fontweight='bold'
            )

        line_min = min(xmin, ymin)
        line_max = max(xmax, ymax)
        xline = np.array([line_min, line_max])

        ax.plot(xline, xline, linestyle='--', color='gray', linewidth=1.5)

        if sd > 0:
            ax.plot(xline, xline - (z_thresh * sd), linestyle=':', color='black', linewidth=1)
            ax.plot(xline, xline + (z_thresh * sd), linestyle=':', color='black', linewidth=1)
        else:
            ax.plot(xline, xline, linestyle=':', color='black', linewidth=1)

        ax.set_xlim(xmin, xmax)
        ax.set_ylim(ymin, ymax)
        ax.set_xlabel(xlabel, fontsize=11)
        ax.set_ylabel(ylabel, fontsize=11)

    # ====== Create all three plots ======
    fig, axes = plt.subplots(1, 3, figsize=(21, 6))

    make_plot(
        ax=axes[0],
        x_col='Density_PATERNAL', y_col='Density_MATERNAL',
        outlier_col='is_outlier_density',
        sd=sd_dens,
        xlabel='Paternal Density', ylabel='Maternal Density',
        xmin=xmin_density, xmax=xmax_density,
        ymin=ymin_density, ymax=ymax_density
    )
    axes[0].set_title('Density', fontsize=13, fontweight='bold')

    make_plot(
        ax=axes[1],
        x_col='Total_Length_PATERNAL', y_col='Total_Length_MATERNAL',
        outlier_col='is_outlier_length',
        sd=sd_len,
        xlabel='Paternal Desired Read Length (bp)', ylabel='Maternal Desired Read Length (bp)',
        xmin=xmin_len, xmax=xmax_len,
        ymin=ymin_len, ymax=ymax_len
    )
    axes[1].set_title('Length', fontsize=13, fontweight='bold')

    make_plot(
        ax=axes[2],
        x_col='Product_PATERNAL', y_col='Product_MATERNAL',
        outlier_col='is_outlier_product',
        sd=sd_prod,
        xlabel='Paternal Density × Length', ylabel='Maternal Density × Length',
        xmin=xmin_product, xmax=xmax_product,
        ymin=ymin_product, ymax=ymax_product
    )
    axes[2].set_title('Density × Length', fontsize=13, fontweight='bold')

    plt.tight_layout()

    if save_path:
        fig.savefig(save_path, format='svg', bbox_inches='tight')

    plt.show()


# ====== Call the function ======
plot_cenpa_dosage(df_combined=CENPA_CDR_A_combined_young_df, df_cdr=CENPA_CDR_young_df,
                  xmin_density=0, xmax_density=0.025,
                  ymin_density=0, ymax_density=0.025,
                  xmin_len=0,     xmax_len=150000,
                  ymin_len=0,     ymax_len=150000,
                  xmin_product=0, xmax_product=4000,
                  ymin_product=0, ymax_product=4000,
                  save_path='/private/groups/migalab/dan/CENPA_young.svg')

plot_cenpa_dosage(df_combined=CENPA_CDR_A_combined_old_df, df_cdr=CENPA_CDR_old_df,
                  xmin_density=0, xmax_density=0.03,
                  ymin_density=0, ymax_density=0.03,
                  xmin_len=0,     xmax_len=150000,
                  ymin_len=0,     ymax_len=150000,
                  xmin_product=0, xmax_product=4000,
                  ymin_product=0, ymax_product=4000,
                  save_path='/private/groups/migalab/dan/CENPA_old.svg')

plot_cenpa_dosage(df_combined=CENPC_CDR_A_combined_young_df, df_cdr=CENPC_CDR_young_df,
                  xmin_density=0, xmax_density=0.055,
                  ymin_density=0, ymax_density=0.055,
                  xmin_len=0,     xmax_len=150000,
                  ymin_len=0,     ymax_len=150000,
                  xmin_product=0, xmax_product=6000,
                  ymin_product=0, ymax_product=6000,
                  save_path='/private/groups/migalab/dan/CENPC_young.svg')

plot_cenpa_dosage(df_combined=CENPA_CDR_A_combined_iPSC_df, df_cdr=CENPA_CDR_iPSC_df,
                  xmin_density=0,    xmax_density=0.01,
                  ymin_density=0,    ymax_density=0.01,
                  xmin_len=0,        xmax_len=300000,
                  ymin_len=0,        ymax_len=300000,
                  xmin_product=0,    xmax_product=1500,
                  ymin_product=0,    ymax_product=1500,
                  save_path='/private/groups/migalab/dan/CENPA_iPSC.svg')
````
